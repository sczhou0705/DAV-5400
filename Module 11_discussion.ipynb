```python

filename1 = "https://raw.githubusercontent.com/sczhou0705/DAV-5400/main/bridges.data1.csv"
df2 = pd.read_csv(filename1)
df2
```

```python
# convert column"Length" to numeric type in case error "No numeric types to aggregate"
df2['Length'] = df2['Length'].apply(pd.to_numeric, errors='coerce')
df2
```

```python
# groupby purpose and material columns and assign it to grouped.
grouped = df2.groupby(['Purpose','Material'])

grouped_arrlen = grouped['Length']
df3 = pd.DataFrame(grouped_arrlen.agg(['mean']))
df3.rename(columns = {'mean':'Average Length'},inplace = True)
df3
```

```python
# Task 3
# split the Distance column into 4 separate bins (one for erected year)
Erected =pd.cut(df2.Erected,4,precision =0)
Erected
```

```python
## define a function that specifies the aggregating functions we want to apply
# to the 'Length' values contained within each of our Dist binning cuts
def get_stats(group):
    return{'Average Length':group.mean(),'Count':group.count(),'Max Length':group.max(),'Min Length':group.min()}

```

```python
#group the df2 column's values relative to the Dist binning cuts we made
# above
grouped = df2.Length.groupby(Erected)
# apply the get_stats function to the grouped data2 values
# In the results we see the data1 quartile cuts are used as the row index labels
grouped.apply(get_stats).unstack()
```
